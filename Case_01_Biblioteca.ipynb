```python
import pandas as pd
```

Carregando a nova base:

```python
aluguel_casa = pd.read_csv("/content/houses_rent.csv")
```

```python
aluguel_casa.head(10)
```

Vamos agora analisar as 5 ultimas linhas do dataset:

```python
aluguel_casa.tail(10)
```

Agora vamos alterar uma coluna para estruturar o dado tirando os espaços na nomenclatura das colunas:

```python
aluguel_casa = aluguel_casa.rename(columns={'parking spaces':'parking_spaces', 'rent amount':'rent_amount', 'property tax':'property_tax', 'fire insurance':'fire_insurance'})
```

```python
aluguel_casa.head()
```

Constinaundo com a limpeza do dataset agora vamos alterar o type das colunas:

```python
aluguel_casa.dtypes
```

```python
aluguel_casa = aluguel_casa.astype({'rent_amount':'float', 'property_tax':'float', 'fire_insurance':'float', 'total':'float'})
aluguel_casa.dtypes
```

Agora vamos explorar a função info que fornece mais informação do dataset

```python
aluguel_casa.info()
```

```python
aluguel_casa[['area', 'rooms', 'bathroom', 'parking_spaces', 'rent_amount', 'property_tax', 'fire_insurance', 'total']]
```

Agora, eu quero ver as pricipais estatísticas de cada coluna. Aqui começamos uma análise descritiva.

```python
aluguel_casa[['area', 'rooms', 'bathroom', 'parking_spaces', 'rent_amount', 'property_tax', 'fire_insurance', 'total']].describe()
```

Agora vamos adentrar um pouco nos filtros partindo de uma premissa que de negócio onde eu posso pagar até 4000 de aluguel:

```python
aluguel_casa = aluguel_casa[(aluguel_casa['total']<=4000)]
aluguel_casa.head()
```

```python
aluguel_casa.describe()
```

Agora além dos 4000 eu preciso de um filtro que também aceite animal:

```python
aluguel_casa = aluguel_casa[(aluguel_casa['total']<=4000) & (aluguel_casa['animal']=='accept')]
aluguel_casa.head()
```

Agora com o filtro de 2 OU 3 quartos:

```python
aluguel_casa = aluguel_casa[(aluguel_casa['total']<=4000) & (aluguel_casa['animal']=='accept') & (aluguel_casa['rooms']==2)|(aluguel_casa['rooms']==3)]
aluguel_casa.head()
```

```python
aluguel_casa.describe()
```

Outra forma de fazer filtros podemos usar o função .query:

```python
aluguel_casa = aluguel_casa.query(("total <= 4000 and animal == 'accept' and rooms == 2 or rooms == 3"))
aluguel_casa.head()
```

Agora vamos verificar a função de agrupamento ou agregação e o reset_index para formatação:

```python
agg_city_price = aluguel_casa.groupby(['city'])['total'].agg('mean').reset_index()
agg_city_price.head()
```

Agora eu quero agrupar por qtd de quartos e banheiros:

```python
agg_bath_room_price = aluguel_casa.groupby(['bathroom', 'rooms'])['total'].agg('mean').reset_index()
agg_bath_room_price.head()
```

Agora vamos explorar a bibliotexa SEABORN para vizialização de dados:

```python
import seaborn as sns
```

Abaixo geramos um gráfico definando os eixos, escolhendo a cor das barras, alterando o nome dos eixos e usando o ; no final para retirar o código da parte de cima do gráfico:

```python
ax = sns.barplot(agg_bath_room_price, x='rooms', y='total', color='blue');
ax.set(xlabel = 'qtd quartos', ylabel = 'média do preço');
```

```python
ax = sns.barplot(agg_bath_room_price, x='bathroom', y='total', color='green');
ax.set(xlabel = 'qtd banheiros', ylabel = 'média do preço');
```

Aqui agrupamos a quantidade de apartamentos por andar e usamos o index para trazer essa quantidade por linha:
