# Contexto:

O banco que solicitou o modelo de limite de crédito incluiu novas variáveis no dataset como escolaridade, gênero e estado do cliente.

Faça as transformações necessárias nessas variáveis Dummies e tire o modelo OLS até que o modelo fique ajustado.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels.formula.api as snf
```

```python
df = pd.read_excel('/content/Limite_Credito_Dummy.xlsx')
```

```python
df.head()
```

```python
# Crie um dicionário com os mapeamentos de nomes antigos para novos
rename_dict = {
    'Gênero': 'Região',
    'Região': 'Gênero',
}

# Use o método rename passando o dicionário criado
df.rename(columns=rename_dict, inplace=True)

# Verifique o DataFrame para confirmar a mudança
df.head()
```

```python
df.corr()
```

```python
df = pd.get_dummies(df, prefix=['Escolaridade', 'Região', 'Gênero'])
df.head()
```

```python
df.corr()
```

```python
function = "LimitedoChequeEspecial~Idade+RendimentoTotal+Salario+LimitedeCreditoImediato+Escolaridade_Fundamental_Medio+Escolaridade_Superior_Pos+Região_RJ+Região_SP+Gênero_F+Gênero_M"
model = snf.ols(formula=function, data=df).fit()
print(model.summary())
```
