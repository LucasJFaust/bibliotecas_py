```python
import pandas as pd
import seaborn as sns
```

```python
vendas = pd.read_excel("/content/varejo.xlsx")
vendas.head()
```

Usando a função info para obter mais informações sobre o dataset. Foi possível verificar que algumas colunas possuem valores nulos que deverão ser tratados.

```python
vendas.info()
```

Aqui começamos a investigar as colunas de maneira mais detalhada. Para isso fizemos um groupby com a adição da função nunique com uma chave (idcompra) onde foi possivel observar que APP e aplicativo representam o mesmo tipo de dado, logo iremos normalizar a informação:

```python
vendas.groupby('idcanalvenda').idcompra.nunique()
```

Utilizamos a função str.replace para realizar a mudança:

```python
vendas['idcanalvenda'] = vendas['idcanalvenda'].str.replace('APP', 'Aplicativo')
vendas.groupby('idcanalvenda').idcompra.nunique()
```

```python
vendas.groupby('bandeira').idcompra.nunique()
```

Aqui vamos seguir a boa prática de colocar o _ entre os espaços:

```python
vendas.groupby('Nome_Departamento').idcompra.nunique()
```

```python
vendas.groupby('estado').idcompra.nunique()
```

```python
vendas['Nome_Departamento'] = vendas['Nome_Departamento'].str.replace(' ', '_')
vendas.groupby('Nome_Departamento').idcompra.nunique()
```

Agora vamos tratar os dado nulls:

```python
vendas.info()
```

```python
vendas[vendas['estado'].isnull()]
```

Para tratar os dados nulos vamos criar uma nova base usando a função drop para deletar os valores nulos:



```python
vendas_sem_nulos = vendas.dropna()
vendas_sem_nulos.info()
```

Atendendo aos requisitos de negócio do case que diz que temos um erro de sistema e que para compras sem UF, devemos considerar o estado de MS. Para isso vamos usar uma função que preenche valores nulos fillna e dessa vez economizamos código usando inplace=True que subsitui direto o valor sem preisar citar mais uma vez a coluna:

```python
vendas['estado'].fillna('MS', inplace=True)
```

```python
vendas.info()
```

Agora vamos tratar o valores nulos da coluna preço. Uma das formas de preencher esses dados é através da média dos valores:

```python
media_preco = vendas['Preço'].mean()
vendas['Preço'].fillna(media_preco, inplace= True)
```

```python
media_preco
```

```python
vendas.info()
```

```python
vendas[['Preço', 'Preço_com_frete']].describe()
```

A regra de negócio diz que o preço não pode ser maior que o preço com frete. Neste caso vamos identificar e eliminar essas linhas:

```python
preco_errado = vendas.query('Preço > Preço_com_frete')

vendas_correto = vendas.query('Preço < Preço_com_frete')
```

```python
vendas_correto.describe()
```

```python
vendas_correto.head()
```

```python
vendas_correto.query("Nome_Departamento == 'Esporte_e_Lazer' and estado == 'SP'").describe()
```

Agora vamos responder algumas das métricas de negócio.
1. Departamento mais vendido
2. Média de preço com frete por nome de Departamento
3. Quantidade de vendas por mês

```python
vendas_correto.groupby('Nome_Departamento').idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
round(vendas_correto.groupby('Nome_Departamento')['Preço_com_frete'].agg('mean').sort_values(ascending=False).reset_index(),2)
```

```python
vendas_correto.groupby('Data').idcompra.nunique().sort_values(ascending=False).reset_index()
```

Agora vamos criar uma coluna para agruparmos o dado por mês:

```python
vendas_correto['mes'] = vendas_correto['Data'].dt.month
vendas_correto.head()
```

```python
vendas_correto.groupby('mes').idcompra.nunique().sort_values(ascending=False).reset_index()
```

Agora vamos explorar a nossa segunda base:

```python
cliente = pd.read_excel('/content/cliente_varejo.xlsx')
cliente.head()
```

```python
cliente[['idade', 'renda']].describe()
```

```python
cliente.info()
```

```python
cliente = cliente.astype({'renda':'float'})
cliente.info()
```

Agora vamos realizar o join entre as tabelas:

```python
vendas_clientes = vendas_correto.merge(cliente, how= 'left', on='cliente_Log')
vendas_clientes.head()
```

Agora vamos responder as duas últimas métricas:
4. Média renda para cada tipo de canal de venda
5. Média de idade de clientes por bandeira

```python
agg_idcanal_renda = round(vendas_clientes.groupby('idcanalvenda')['renda'].agg('mean').sort_values(ascending=False).reset_index(),2)
agg_idcanal_renda
```

```python
agg_idade_bandeira = round(vendas_clientes.groupby('bandeira')['idade'].agg('mean').sort_values(ascending=False).reset_index(),2)
agg_idade_bandeira
```

Agora vamos trabalhar com visualizações:

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(10,8))

plt.bar(agg_idade_bandeira['bandeira'], agg_idade_bandeira['idade'], color='green')

plt.ylabel('Média de idade')

plt.title('Idade média por Bandeira', loc='center', fontsize=18, fontweight='bold', color='black')
```

```python
fig = plt.figure(figsize=(7,4))

plt.bar(agg_idcanal_renda['idcanalvenda'], agg_idcanal_renda['renda'], color='green')

plt.ylabel('Renda média')

plt.title('Renda média por canal de venda', loc='center', fontsize=18, fontweight='bold', color='black')
```

```python
vendas_por_data = vendas_correto.groupby('Data').idcompra.nunique().reset_index()
vendas_por_data
```

Agora vamos plotar um gráfico de linha

```python
fig = plt.figure(figsize=(20,4))

plt.xlabel('Data da Venda')
plt.ylabel('Quantidade de Vendas')
x = vendas_por_data['Data']
y = vendas_por_data['idcompra']

plt.plot(x,y, color='green', linewidth='2')
plt.show()
```

Agora vamos criar um gráfico para média de preço por departamento:

```python
agg_debt_preco = round(vendas_correto.groupby('Nome_Departamento')['Preço_com_frete'].agg('mean').sort_values(ascending=False).reset_index(),2)
agg_debt_preco
```

```python
fig = plt.figure(figsize=(40,20))

plt.bar(agg_debt_preco['Nome_Departamento'], agg_debt_preco['Preço_com_frete'], color='green')

plt.ylabel('Renda média')

plt.title('Preço médio por departamento', loc='center', fontsize=18, fontweight='bold', color='black')
```
