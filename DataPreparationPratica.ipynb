Importando as bibliotécas que vamos usar para tratar os dados

```python
import pandas as pd
import numpy as np
```

Aqui estou criando um DataFrame com a base carregada e depois uso um função para ler esse DataFrame criado

```python
df = pd.read_csv("DataPrepFinal.csv")
```

```python
df.head()
```

Agora estou definindo um novo DataFrame ja com o uso do drop para excluir colunas desnecessárias.

```python
df = df.drop(columns=['Unnamed: 0', 'name', 'category', 'goal', 'pledged', 'usd pledged'])
```

```python
df.head()
```

Aqui estamos usando uma função para analisar os tipos dos dados. E possível verificar que a coluna usd_goal_real é um objeto do tipo string quando na realidade deveria ser um valor numérico int64.


```python
df.dtypes
```

Agora vamos aplicar uma função apply onde eu executo alguma coisa em todas as linhas da coluna. Como queremos retirar de todas as linhas o 'USD', vamos aplicar a função do pandas chamada de strip através do lambda que sinaliza que vamos escrever a nossa propria função pois não existe uma função especifica para o que vamos fazer.

```python
df['usd_goal_real'] = df['usd_goal_real'].apply(lambda x: x.strip('USD'))
```

```python
df.head()
```

Agora vamos fazer um string replace (str.replace) do pandas associado com um Regex que é uma forma bem útil e eficiente para coletar informações em strings, encontrar padrões de textos no Python. Vamos fazer isso para coluna de launched, pois diferente da coluna anterior os valores depois de datas são diferentes. Poderiamos fazer essa mudança com funções datetime também.

```python
df['launched'] = df['launched'].str.replace(' \d\d:\d\d', '', regex=True)
```

```python
df.head()
```

Agora vamos mudar os tipo das valores na coluna.

```python
df['usd_goal_real'] = df['usd_goal_real'].astype('int64')
```

```python
df.dtypes
```

Notei que cometi um erro ao fazer o regez e acabei adicionando uma coluna launched\t na tabela. Agora vou usar o método drop com inplace=True para excluí-la da tabela diretamente no DataFrame sem ter que reatribuir.

```python
df.drop('launched\t', axis=1, inplace=True)
```

```python
df.head()
```

Agora vamos transformar uma coluna em datetime com a função do pandas. Lembrando que temos que especificar o formato devido as particularidades dos pandas e das diferêncas em relação a cada região de data e hora.

```python
df['deadline'] = pd.to_datetime(df['deadline'], format='%d/%m/%y')
df['launched'] = pd.to_datetime(df['launched'], format='%d/%m/%y')
```

```python
df.head()
```

```python
df.dtypes
```

Agora vamos construir um dado novo, ou seja, uma task de construção que indica o tempo que ficou aberto essa campanha. Tempo de lançamento e tempo deadline.

```python
df['time_range'] = (df['deadline']-df['launched']).dt.days
```

```python
df.head()
```

Agora nesse merge temos que ter cuidado, vamos fazer um group by para vizualizar como ele não conseguiu resolver.

Para resolver essa questão temos que aplicar a ténica de One Hoting Coding, onde vamos pegar toda a variável categorica e transformar em variável boleana, ou seja, vou criar uma coluna, por exemplo, com USA e vou colocá-la como TRUE ou valor 1 para todo mundo que ter o atributo de USA. Ai vamos fazer isso para as outras categorias.

```python
df_invest.head()
```

```python
df_invest = pd.get_dummies(df_invest, columns= ['backedLocation'])
```

```python
df_invest.head()
```

```python
df_invest.sort_values(by=['ID'])
```

Agora faz sentindo juntar os ID iguas em uma linha com o group by e uma agregação de soma investimentos de informações com média de idade

```python
df_invest = df_invest.groupby(by=['ID']).agg({'age':'mean', 'backedLocation_BR':'sum', 'backedLocation_GBK':'sum', 'backedLocation_US':'sum'})
```

```python
df_invest.head()
```

Agora vamos dar um merge do novo dataset invest no dataset principal

```python
df = df.merge(df_invest, how='right', on='ID')
```
