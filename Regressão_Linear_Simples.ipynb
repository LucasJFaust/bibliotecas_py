```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
dataset = {"tempo":[1,3,4,4,6,8,10,10,11,13],
           "vendas":[91,110,106,116,119,129,139,143,138,159]
}

dataset = pd.DataFrame(dataset)
dataset.head()
```

```python
ax = sns.lmplot(x="tempo", y="vendas", data=dataset)
ax.fig.set_size_inches(12,6)
ax.fig.suptitle("Reta de Regressão - Tempo X Vendas", fontsize=16, y=1.02)
ax.set_xlabels("Tempo de Experiência", fontsize=14)
ax.set_xlabels("Vendas", fontsize=14)
ax;
```

```python
dataset.corr()
```

```python
import statsmodels.api as sm
```

```python
Y = dataset.vendas
X = sm.add_constant(dataset.tempo)
```

```python
resultado_regressao = sm.OLS(Y,X).fit()
```

```python
print(resultado_regressao.summary())
```

```python
dataset["Y_previsto"] = resultado_regressao.predict()
dataset.head()
```

```python
dataset["residuo"] = dataset["Y_previsto"] - dataset["vendas"]
dataset.head()
```

```python
dataset.drop('Y_previsto)', axis=1, inplace=True)
dataset.head()
```
